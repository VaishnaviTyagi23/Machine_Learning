**BASICS**


*   Machine Learning is a subset of Artificial Intelligence.
*   Deep Learning is a subset of Machine Learning.



**AI**:
Artificial Intelligence is a branch of computer science that is concerned with building smart and intelligent machines.

eg. Non-Intelligent machines: Bike,car
eg. Intelligent machines: Google Assistant, Tesla.

**Machine Learning:** It is a technique to implement AI that can learn from the data by themselves without being explicitly programmed.

**Deep Learning:**It is a subfield of Machine Learning that uses Artificial Neural Networks to learn from the data .
It has many layers in ANN.

**Types of ML:**


1.   Supervised Learning: the ML algorithm learns from labelled data.
2.   Unsupervised Learning: the ML algorithm learns from unlabelled data.
3.   Reinforcement Learning: is an area of ML concerned with how intelligent agents take actions in an environment to maximize its rewards.
a)  Environment
b)  Agent
c)  Action
d)  Reward





**Types of Supervised Learning:**
The ML algorithm learns from labelled data.

1.   Classification: It is about predicting a class or discrete values Eg. Male or Female ;True or False
2.   Regression: It is about predicting a quantity or continuous values. Eg. Salary; age; price



Algorithms for Classification are:

1.   Decision Tree
2.   Random Forest
3.   KNN

Algorithms for Regression are:

1.   Logistic Regression
2.   Polynomial Regression
3.   SVM




**Types of Unsupervised Learning:**
The ML algorithm learns from unlabelled data.

1.   Clustering: It is unsupervised task
 wich involves grouping the similar data points.Eg. mobile network company project-> divide into 2 clusters with people who use net more, people who use call more
2.   Association: It is unsupervised task that is used to find important relationship between data points. Eg. if someone buys bread they most probably buy milk; OTT platform recommendation.


Algorithms for Unsupervised Learning Algorithms are:

1.   KNN
2.   Hierarchical Clustering
3.   PCA(Principal Component Analysis)
4.   Apriori
5.   Eclat
